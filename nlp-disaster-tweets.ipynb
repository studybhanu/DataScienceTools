```python
def get_glove_embeddings_from_file(path):
    glove = open(path,"r",encoding = "utf-8")
    glove_dict ={}
    for line in glove:
        values = line.split()
        word =values[0]
        word_vec = np.asarray(values[1:])
        glove_dict[word] = word_vec
    glove.close()
    return glove_dict
```

```python
def get_glove_embeddings(tokenizer,glove_dict,word_length):
    vocab = len(tokenizer.word_index)+1
    embedding_matrix = np.zeros((vocab,word_length))
    for word , i in tokenizer.word_index.items():
        embedding_vector = glove_dict.get(word)
        if embedding_vector is not None:
            embedding_matrix[i] = embedding_vector
    return embedding_matrix
```

```python

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding,Dense,Flatten,BatchNormalization,LayerNormalization, Bidirectional,Dropout,LSTM,GRU
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping
from tensorflow import keras
```

```python
filepath="weights.best.hdf5"
checkpoint = ModelCheckpoint(filepath, monitor='val_accuracy', verbose=1, save_best_only=True, mode='max')
es = EarlyStopping(monitor='accuracy', patience=3)
callbacks_list = [checkpoint, es]
```

```python
def build_glove_embeddings_text_model(embedding_matrix,vocab,input_length,word_length):
    model = Sequential()
    model.add(Embedding(vocab,word_length,input_length=input_length,weights=[embedding_matrix],trainable =False))
    
    model.add(Bidirectional(GRU(256,return_sequences=True)))
    
    model.add(Bidirectional(GRU(256,return_sequences=True)))
    
    model.add(Bidirectional(GRU(128,return_sequences=True)))
    
    model.add(Bidirectional(GRU(128,return_sequences=True)))

    model.add(Bidirectional(GRU(128,return_sequences=True)))
    
    model.add(Bidirectional(GRU(64,return_sequences=True)))
    
    model.add(Bidirectional(GRU(64,return_sequences=True)))
    
    model.add(Bidirectional(GRU(64,return_sequences=True)))
    
    model.add(Bidirectional(GRU(64,return_sequences=True)))
    
    model.add(Bidirectional(GRU(32,return_sequences=True)))
    
    model.add(Bidirectional(GRU(32,return_sequences=True)))
    
    model.add(Bidirectional(GRU(32,return_sequences=True)))
    
    model.add(Bidirectional(GRU(32,return_sequences=True)))

    model.add(Bidirectional(GRU(32,return_sequences=True)))

    
    
    model.add(Flatten())

    model.add(Dense(1,activation="sigmoid"))
    optimizer = keras.optimizers.Adam(learning_rate=0.0001,clipnorm=1)
    #model.load_weights("weights.best.hdf5")
    model.compile(optimizer = optimizer,loss ="binary_crossentropy",metrics=["accuracy"])
    model.summary()
    return model
```

```python
glove_embedded_model =  build_glove_embeddings_text_model(embedding_matrix,vocab=vocab_size,input_length = 100,word_length=50)
```

```python
glove_embedded_model.fit(X_train,y_train,epochs = 30,verbose =1,batch_size = 5,validation_split = 0.1,validation_data=(X_test,y_test),callbacks=[callbacks_list])
```

```python
import keras_tuner 
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
import keras_tuner as kt
from tensorflow.keras.losses import binary_crossentropy

from kerastuner import HyperModel


# HyperModel is an abstract class that defines the structure of a model that can be tuned.
# it requires the implementation of a build method, which takes in a HyperParameters object and returns a compiled Keras model. This allows the HyperModel to be customized to the specific requirements of the problem being solved.

from kerastuner import HyperParameters

# HyperParameters is a class that provides an interface for defining hyperparameters for a Keras model. It allows the specification of different types of hyperparameters, such as Int, Float, Choice, etc. 
# It also provides methods for sampling hyperparameters, which are used during the hyperparameter search process.
```

```python
# creates an instance of the HyperParameters class from the Keras Tuner (kt) library.

# Hyperparameters are variables that are set before training a machine learning model, and can significantly impact its performance. Examples of hyperparameters include learning rate, batch size, number of layers, number of neurons in each layer, regularization coefficients, and activation functions.
hp = kt.HyperParameters()
```

```python
filepath="weights.best.hdf5"
checkpoint = ModelCheckpoint(filepath, monitor='val_accuracy', verbose=1, save_best_only=True, mode='max')
es = EarlyStopping(monitor='accuracy', patience=3)
callbacks_list = [checkpoint, es]
```

```python


class Tuning(HyperModel):



    def __init__(self, vocab, word_length, input_length,embedding_matrix ,num_classes):
            self.vocab = vocab
            self.word_length = word_length
            self.input_length = input_length
            self.embedding_matrix= embedding_matrix
            self.num_classes = num_classes
           



    def build(self,hp):
        model = keras.Sequential()
        
        model.add(Embedding(
                                input_dim=self.vocab,
                                output_dim=self.word_length,
                                input_length=self.input_length,
                                trainable=False, 
                                weights=[self.embedding_matrix]))        
        
        
        # Tune the number of hidden layers
        for i in range(hp.Int('num_layers', min_value=2, max_value=8)):
            
                    # Add a Bidirectional GRU layer with return_sequences=True
                model.add(Bidirectional(GRU(units=hp.Int('units',
                                                  min_value=128,
                                                  max_value=512,
                                                  step=32),
                                    activation='relu',
                                    return_sequences=True)))
                
                model.add(BatchNormalization())
                # Add dropout regularization
                model.add(layers.Dropout(hp.Float('dropout_' + str(i), min_value=0.0, max_value=0.3, step=0.1)))
                

 




        model.add(Flatten())
    
        # Add a Dense layer
        model.add(Dense(units=self.num_classes-1, activation='sigmoid'))

        # Compile the model
        model.compile(loss='categorical_crossentropy',
                      optimizer=keras.optimizers.Adam(
                          hp.Choice('learning_rate',
                                    values=[1e-3, 1e-2])),
                      metrics=['accuracy'])

        return model
```

```python
tuning1 = Tuning(vocab=vocab_size,word_length=50,input_length = 100,embedding_matrix=embedding_matrix,num_classes=2) # creating an object for Tuning class
```

```python
tuning1.build(hp)
```

```python
from kerastuner.tuners import BayesianOptimization
"""
The arguments to the Hyperband constructor are:

tuning1: an instance of a HyperModel subclass, which defines the machine learning model and its associated hyperparameters that will be tuned by the Hyperband tuner.
objective: an instance of the Objective class from the Keras Tuner library, which specifies the metric to be optimized during the hyperparameter search. In this case, the metric is f1_m, which is a custom metric defined elsewhere in the code.

"""
tuner = kt.BayesianOptimization(
    tuning1,
    objective=kt.Objective('accuracy', direction='max'),
    max_trials = 10
    )
```

```python




"""
Hyperband tuner, using the search method. The search method takes several arguments:

X_train_total and y_train: the training data and labels to be used for training and validation during the hyperparameter search.
epochs: the maximum number of epochs to train each model during the hyperparameter search.
validation_split: the fraction of the training data to use for validation during the hyperparameter search.
batch_size: the batch size to use for training and validation during the hyperparameter search, which is sampled from the hyperparameter space defined by the hp.Int method with minimum value 10, maximum value 40, and step 2.
"""
tuner.search(X_train_total, y_train, epochs=2, validation_split=0.2,batch_size=hp.Int('batch_size', min_value=50, max_value=100, step=10),callbacks=[callbacks_list])
```
