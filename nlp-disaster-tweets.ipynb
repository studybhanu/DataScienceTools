```python
def get_glove_embeddings_from_file(path):
    glove = open(path,"r",encoding = "utf-8")
    glove_dict ={}
    for line in glove:
        values = line.split()
        word =values[0]
        word_vec = np.asarray(values[1:])
        glove_dict[word] = word_vec
    glove.close()
    return glove_dict
```

```python
def get_glove_embeddings(tokenizer,glove_dict,word_length):
    vocab = len(tokenizer.word_index)+1
    embedding_matrix = np.zeros((vocab,word_length))
    for word , i in tokenizer.word_index.items():
        embedding_vector = glove_dict.get(word)
        if embedding_vector is not None:
            embedding_matrix[i] = embedding_vector
    return embedding_matrix
```

```python
embedding_matrix=get_glove_embeddings(tokenizer = t,glove_dict = glove_dict2,word_length=50)
```

```python

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding,Dense,Flatten,BatchNormalization,LayerNormalization, Bidirectional,Dropout,LSTM,GRU
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping
from tensorflow import keras
```

```python
filepath="weights.best.hdf5"
checkpoint = ModelCheckpoint(filepath, monitor='val_accuracy', verbose=1, save_best_only=True, mode='max')
es = EarlyStopping(monitor='val_accuracy', patience=3)
callbacks_list = [checkpoint, es]
```

```python
def build_glove_embeddings_text_model(embedding_matrix,vocab,input_length,word_length):
    model = Sequential()
    model.add(Embedding(vocab,word_length,input_length=input_length,weights=[embedding_matrix],trainable =False))
    
    model.add(Bidirectional(GRU(128,return_sequences=True)))
    model.add(Dropout(0.1))
    model.add(Bidirectional(GRU(128,return_sequences=True)))
    model.add(Dropout(0.1))
    model.add(Bidirectional(GRU(128,return_sequences=True)))
    model.add(Dropout(0.1))
    model.add(Bidirectional(GRU(128,return_sequences=True)))
    model.add(Dropout(0.1))
    model.add(Bidirectional(GRU(128,return_sequences=True)))
    model.add(Dropout(0.1))
    model.add(Bidirectional(GRU(128,return_sequences=True)))
    model.add(Dropout(0.1))
    model.add(Bidirectional(GRU(128,return_sequences=True)))
    model.add(Dropout(0.1))
    model.add(Bidirectional(GRU(128,return_sequences=True)))
    model.add(Dropout(0.1))
    model.add(Bidirectional(GRU(128,return_sequences=True)))
    model.add(Dropout(0.1))
    model.add(Bidirectional(GRU(128,return_sequences=True)))
    model.add(Dropout(0.1))
    model.add(Bidirectional(GRU(128,return_sequences=True)))
    model.add(Dropout(0.1))
    model.add(Bidirectional(GRU(128,return_sequences=True)))
    model.add(Dropout(0.1))
    model.add(Bidirectional(GRU(128,return_sequences=True)))
    model.add(Dropout(0.1))
    model.add(Bidirectional(GRU(128,return_sequences=True)))

    
    
    model.add(Flatten())

    model.add(Dense(1,activation="sigmoid"))
    optimizer = keras.optimizers.Adam(learning_rate=0.0001,clipnorm=1)
    #model.load_weights("weights.best.hdf5")
    model.compile(optimizer = optimizer,loss ="binary_crossentropy",metrics=["accuracy"])
    model.summary()
    return model
```

```python
glove_embedded_model =  build_glove_embeddings_text_model(embedding_matrix,vocab=vocab_size,input_length = 100,word_length=50)
```

```python
glove_embedded_model.fit(X_train,y_train,epochs = 10,verbose =1,batch_size = 5,validation_split = 0.2,validation_data=(X_train,y_train),callbacks=[callbacks_list])
```

```python
# evaluate model 
loss, accuracy =glove_embedded_model.evaluate(X_train,y_train,verbose =1)
print("accuracy :%f" %(accuracy*100))
```

```python
# evaluate model 
loss, accuracy =glove_embedded_model.evaluate(X_test,y_test,verbose =1)
print("accuracy :%f" %(accuracy*100))
```
